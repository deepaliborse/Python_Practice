```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars=pd.read_csv("Toyota.csv", index_col=0,na_values=['??','????'])
cars.shape
cars.head()
```

```python
cars.dropna(axis=0, inplace=True)
cars.shape
```

# Scatter plot

```python
plt.scatter(cars['Age'], cars['Price'], c="green")

plt.title("Price vs Age")
plt.xlabel("Price")
plt.ylabel("Age")
```

# Histogram
- Used for frequency distribution

```python
plt.hist(cars['KM'])
```

```python
plt.hist(cars['KM'],
        color='maroon',
        edgecolor='white',
        bins=10)

plt.title("Histogram of KM-frequency")
plt.xlabel("KM")
plt.ylabel("Frequency")
```

```python
plt.hist(cars['KM'],
        color='maroon',
        edgecolor='white',
        bins=10,
        orientation='horizontal')

plt.title("Histogram of KM-frequency")
plt.xlabel("KM")
plt.ylabel("Frequency")
```

# Bar plots

```python
cars['FuelType'].value_counts()
```

```python
FuelTypes=['Petrol','Diesel','CNG']
counts=[968,116,12]
index=np.arange(len(FuelTypes))
```

```python
plt.bar(FuelTypes,counts, color=['lightsteelblue','cornflowerblue','darkblue'])

plt.title("Histogram of FuelType-frequency")
plt.xlabel("Fueltype")
plt.ylabel("count")


                                                                        #color=['khaki','goldenrod','darkgoldenrod']
```

```python
plt.barh(FuelTypes,counts, color=['lightsteelblue','cornflowerblue','darkblue'])

plt.title("Histogram of FuelType-frequency")
plt.xlabel("Fueltype")
plt.ylabel("count")


                                                                        #color=['khaki','goldenrod','darkgoldenrod']
```

# Lmplot and Hue

```python
sns.lmplot(data=cars,x='Age',y='Price',palette='rocket',hue='FuelType',fit_reg=False)
```

```python
sns.lmplot(data=cars,x='Age',y='Price',palette='icefire',hue='FuelType')
```

# Distplot
- (Distribution plot) with density estimate line

```python
sns.distplot(cars['Age'], color='darkgreen')
```

```python
sns.distplot(cars['Age'], color='brown', kde=False)   #kde=Kernel density estimation
```

```python
sns.distplot(cars['Age'],bins=10, color='darkblue')
```
